```python
import numpy as np
from matplotlib import pyplot as plt
from functools import reduce

%matplotlib inline
```

```python
n = 15
x = np.linspace(0, 8, 20)
y = x + 2*np.sin(1.5*x) + np.random.normal(0, 0.3, 20)

x2 = np.linspace(0,9,300)
f_true = x2 + 2*np.sin(1.5*x2)
```

```python
fig, ax = plt.subplots(1,1, figsize=(10,8))
ax.scatter(x, y, s=15)
ax.plot(x2, f_true, linewidth=1, alpha=.4);
```

```python
ix = np.random.choice([False, True], size=n, replace=True, p=(1/3, 2/3))
x_train, y_train = x[ix], y[ix]
x_test, y_test = x[~ix], y[~ix]

model1 = np.polyfit(x_train, y_train, deg=1)
model1_fit = np.sum(c*(x2**i) for i, c in enumerate(model1[::-1]))

model2 = np.polyfit(x_train, y_train, deg=8)
model2_fit = np.sum(c*(x2**i) for i, c in enumerate(model2[::-1]))
```

```python
fig, ax = plt.subplots(1,1, figsize=(10,8))
ax.scatter(x_train, y_train, s=15)
ax.scatter(x_test, y_test, s=15, c='red')
ax.plot(x2, model1_fit, linewidth=1, alpha=.4);
ax.plot(x2, model2_fit, linewidth=1, alpha=.4);
```
